<h3> <b>Concatenate Data</h3>
We will create one large dataframe consisting of every month of the year. This will make our data easier to analyze.

```python
# Create a list of our dataframes
df_list = [df_1,df_2,df_3,df_4,df_5,df_6,df_7,df_8,df_9,df_10,df_11,df_12,]

# Use pandas concat function to create a single dataframe.
df_2021 = pd.concat(df_list)
```

<h3><b>Transform Data</b></h3>  
    Create new features that will add depth to our analysis.

```python
# Subtract the times to get an amount of time on bike
x = (df_2021['ended_at'].subtract(df_2021['started_at']))
# Extract only the minutes from timedelta dtype
y = x.astype('timedelta64[m]')
#Create a new column called ride_lenght_minutes
df_2021['ride_length_minutes'] = y.abs()
```

* <b>We created a new column('ride_length_minutes) that shows how many minutes our riders used the bike

```python
# Create a new column that shows day of the week as an int.
df_2021['day_of_week'] = df_2021['started_at'].dt.day_name()
```

* <b>We created a new column that shows the day of the week.

```python
df_2021.head()
```

<h3><b>Look for Human Error and Outliers

```python
# Get statistical data on our features
df_2021.describe().round(2)
```

```python
# Create a boxplot to show outliers in ride time
sns.boxplot(x='rider_class', y='ride_length_minutes', data=df_2021)
```

```python
df_2021[df_2021['ride_length_minutes'] <= 100].info()
```

* <b>It seems that there are thousands of entries with a ride time is greater than 12 hours. As a precaution I will drop any entry that has a ride_time_minutes greater than or equal to 100, since that is closer to the ride_time_minutes mean.

```python
x = df_2021['ride_length_minutes'] >= 100
df_2021.drop(df_2021[x].index, inplace=True)
df_2021.info()
```

```python
y = df_2021['ride_length_minutes'] == 0
df_2021.drop(df_2021[y].index, inplace=True)
```

* <b>Drop rides that have a ride length of 0.

```python
sns.boxplot(x='rider_class', y='ride_length_minutes', data=df_2021)
```
